```python
########### PLOTS :)
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# specify paths

# read data from file
file_path_1 = '/home/elenaespinosa/OneDrive/Uni/Summer_courses/Summer_project/mod_lammps_code/outfiles/outfile_1'
file_path_2 = '/home/elenaespinosa/OneDrive/Uni/Summer_courses/Summer_project/mod_lammps_code/outfiles/outfile_2'
file_path_3 = '/home/elenaespinosa/OneDrive/Uni/Summer_courses/Summer_project/mod_lammps_code/outfiles/outfile_3'

# save plots to
save_plots_to = '/home/elenaespinosa/OneDrive/Uni/Summer_courses/Summer_project/mod_lammps_code/plots/'
```

```python
# parse data
data_1 = pd.read_csv(file_path_1, sep=' ', comment='#', header=None)
data_1.columns = ['Timesteps', 'No_of_clusters', 'Mean_size_of_clusters', 'Size_of_largest_cluster']

data_2 = pd.read_csv(file_path_2, sep=' ', comment='#', header=None)
data_2.columns = ['Timesteps', 'No_of_clusters', 'Mean_size_of_clusters', 'Size_of_largest_cluster']

data_3 = pd.read_csv(file_path_3, sep=' ', comment='#', header=None)
data_3.columns = ['Timesteps', 'No_of_clusters', 'Mean_size_of_clusters', 'Size_of_largest_cluster']
```

```python
plt.figure(figsize=(16, 10))

plt.plot(data_1['Timesteps'], data_1['No_of_clusters'], marker='.', alpha=0.7, label='Model 1')
plt.plot(data_2['Timesteps'], data_2['No_of_clusters'], marker='.', alpha=0.7, label='Model 2')
plt.plot(data_3['Timesteps'], data_3['No_of_clusters'], marker='.', alpha=0.7, label='Model 3')

plt.xlabel('Timesteps', fontsize ='16')
plt.xticks(fontsize='14')

plt.ylabel('Number of Clusters', fontsize ='16')
plt.yticks(fontsize='14')

plt.title('Number of Clusters vs Timesteps', fontsize ='16')
plt.ticklabel_format(style='plain')

plt.legend(fontsize="14", loc ="upper right")
plt.grid(True)

plt.savefig(save_plots_to + "plot_1.png")
plt.show()
```

```python
plt.figure(figsize=(16, 10))

plt.plot(data_1['Timesteps'], data_1['Mean_size_of_clusters'], marker='.', alpha=0.7, label='Model 1')
plt.plot(data_2['Timesteps'], data_2['Mean_size_of_clusters'], marker='.', alpha=0.7, label='Model 2')
plt.plot(data_3['Timesteps'], data_3['Mean_size_of_clusters'], marker='.', alpha=0.7, label='Model 3')

plt.xlabel('Timesteps', fontsize ='16')
plt.xticks(fontsize='14')

plt.ylabel('Mean size of clusters', fontsize ='16')
plt.yticks(fontsize='14')

plt.title('Mean size of clusters vs Timesteps', fontsize ='16')
plt.ticklabel_format(style='plain')

plt.legend(fontsize="14", loc ="upper right")
plt.grid(True)

plt.savefig(save_plots_to + "plot_2.png")
plt.show()
```

```python
plt.figure(figsize=(16, 10))

plt.plot(data_1['Timesteps'], data_1['Size_of_largest_cluster'], marker='.', alpha=0.7, label='Model 1')
plt.plot(data_2['Timesteps'], data_2['Size_of_largest_cluster'], marker='.', alpha=0.7, label='Model 2')
plt.plot(data_3['Timesteps'], data_3['Size_of_largest_cluster'], marker='.', alpha=0.7, label='Model 3')

plt.xlabel('Timesteps', fontsize ='16')
plt.xticks(fontsize='14')

plt.ylabel('Size of largest cluster', fontsize ='16')
plt.yticks(fontsize='14')

plt.title('Size of largest cluster vs Timesteps', fontsize ='16')
plt.ticklabel_format(style='plain')

plt.legend(fontsize="14", loc ="upper right")
plt.grid(True)

plt.savefig(save_plots_to + "plot_3.png")
plt.show()
```
