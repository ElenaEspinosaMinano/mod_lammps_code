```python
# dictionary to store data frames
data_frames = {}

# parse data
for i in range(1, 9):
    data_frames[i] = pd.read_csv(path_to_outfile + outfiles_list[i-1], sep=' ', comment='#', header=None)
    data_frames[i].columns = ['Timesteps', 'No_of_clusters', 'Mean_size_of_clusters', 
                              'Size_of_largest_cluster', 'No_of_clusters_of_size_1']
```
